```python
import fitz  # PyMuPDF
from google.colab import files

uploaded = files.upload()
filename = list(uploaded.keys())[0]

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

resume_text = extract_text_from_pdf(filename)
print("✅ Extracted Resume Text:\n")
print(resume_text[:1000])
```

```python
# 🎯 Target job role
job_title = "Data Engineer"

# 🛠️ Required skills for the job (you can customize)
target_skills = [
    "Python", "SQL", "Apache Spark", "AWS", "Airflow", "ETL", "Data Warehousing",
    "Big Data", "Docker", "Kubernetes", "Data Pipelines", "Pandas", "Linux", "Git"
]
```

```python
import re

def extract_skills_from_resume(text, skill_list):
    found = []
    for skill in skill_list:
        pattern = r'\b' + re.escape(skill.lower()) + r'\b'
        if re.search(pattern, text.lower()):
            found.append(skill)
    return found

found_skills = extract_skills_from_resume(resume_text, target_skills)
missing_skills = list(set(target_skills) - set(found_skills))

print("✔️ Skills Found in Resume:\n", found_skills)
print("\n❌ Skills Missing:\n", missing_skills)
```

```python
match_score = round(len(found_skills) / len(target_skills) * 100, 2)
print(f"\n🔐 Resume–Job Match Score: {match_score}%")
```

```python
def generate_feedback(missing_skills):
    suggestions = []
    for skill in missing_skills:
        suggestions.append(f"✅ Add experience using {skill} in your project or job description.")
    return suggestions

feedback = generate_feedback(missing_skills)

print("\n💡 Suggestions to Improve Resume:\n")
for tip in feedback:
    print("•", tip)
```
